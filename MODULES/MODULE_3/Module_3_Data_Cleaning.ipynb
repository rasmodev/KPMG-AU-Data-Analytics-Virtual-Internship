Import libraries

```python
import pandas as pd
import numpy as np
import datasist as ds
```

```python
# Read the data from the excelfile
df = pd.read_excel("KPMG_MODULE_3_Visualization Data.xlsx")
df.head()
```

```python
# Dataframe shape
df.shape
```

```python
# Check the column information
df.info()
```

The dataset is properly formatted but has several unneccessary columns

```python
# Drop Unnecessary Columns
columns_to_drop = [f"Unnamed: {i}" for i in range(30, 44)]
df = df.drop(columns=columns_to_drop)
df.info()
```

```python
df.isnull().sum()
```

```python
# Drop rows with missing values
df = df.dropna()

# Confirm if missing values have been handled
df.isnull().sum()
```

```python
# Check the column information of the cleaned dataset
df.info()
```

```python
#Save cleaned data to a new csv file
df.to_csv('visualization_data.csv', index=False)
```

```python
# Perform EDA with the datasist library
# Summary statistics
summary = ds.structdata.describe(df)
summary
```
